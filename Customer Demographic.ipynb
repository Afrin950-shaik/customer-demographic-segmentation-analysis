```python
bins=[18,25,35,45,55,100]
labels=['18-25','26-35','36-45','46-55','56+']
Customer_data['Age-grouped']=pd.cut(Customer_data["age"],bins=bins,labels=labels)
```

```python
Customer_data['Age-grouped']
```

```python
Customer_data['Income level']=pd.qcut(Customer_data['income'],q=3,labels=['Low','Middle','High'])
```

```python
Customer_data[['Age-grouped','Income level','spending_score']]
```

```python
Customer_data.groupby("Age-grouped")['spending_score'].mean()
```

```python
Customer_data.groupby("Income level")['spending_score'].mean()
```
